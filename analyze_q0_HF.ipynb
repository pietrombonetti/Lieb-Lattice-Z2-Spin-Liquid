```python
Nk = 20

QSL_pars=[]

filename=open('../data/q0_runs/output_complete.txt','w+')

for ii in data[0]:

    ii = int(ii)
    Ui = data[1,ii]
    Dd = data[2,ii]
    Dp = data[3,ii]
    theta = abs(data[6,ii]-data[7,ii])/2
    print('\t %.3f %.3f %.3f %.3f' %(Ui,Dd,Dp,theta))
    if (max(abs(Dd),abs(Dp))>1e-3):
        Dpx = Dp * np.exp(1j*theta)
        Dpy = Dp * np.exp(-1j*theta)
        
        pars = dict(t=1,U=Ui,J=0,t_pxpy=0,t_pp=0,t_dd=0,J_pxpy=0)
        OBJ = HFq0.HF_q0(Nk,pars)
        
        par_ini = dict(mu=0,Delta_d=Dd,Delta_px=Dpx,Delta_py=Dpy)
        T0=0.01
        numpars=dict(T=T0,mix=0.75,Nitermax = 30000,tol = 5e-4,vb=True)
        np.set_printoptions(suppress=True,precision=6)
        print('====================================================')
        print('starting calculation at U = %.2f' %Ui)
        mud,mupx,mupy,Dd,Dpx,Dpy,F,mu = OBJ.SC_loop(n_wanted=0.5,par_ini=par_ini,**numpars,Hartree_term=True,symm2=True)
        datapr = '%.5f  %.5f   %.5f   %.5f   %.5f   %.5f   %.5f   %.5f   %.5f   %.5f   %.10f' %(mud,mupx,mupy,abs(Dd),abs(Dpx),abs(Dpy),np.angle(Dd),np.angle(Dpx),np.angle(Dpy),mu,F)
        print(datapr,file=filename)
        print('====================================================')
        print('calculating QSL parameters')
        
        phase=np.exp(-1j*np.angle(Dd))
        scp = dict(mud=mud,mupx=mupx,mupy=mupx,Dd=Dd*phase,Dpx=Dpx*phase,Dpy=Dpy*phase)
        Ek,phik,Hk = OBJ.make_H(scp)
        mags = OBJ.magnetiz(Ek,phik,T0)
        
        U = np.array([[1-1j,-1-1j],[1-1j,1+1j]]) / 2
        Ur = np.kron(U,np.identity(3))
        Hk = np.einsum('ba,xybc,cd->xyad',np.conj(Ur),Hk,Ur)
        
        sigma_y = 1j*np.array([[0,-1],[1,0]])
        SM = np.kron(sigma_y,np.diag([0,1,-1]))
        theta=-np.angle(mags[1])
        Ur = sp.linalg.expm(-1j*theta/2*SM)
        Hk = np.einsum('ba,xybc,cd->xyad',np.conj(Ur),Hk,Ur)
        
        
        Ek,phik = np.linalg.eigh(Hk)
        _,_,chitau,Qtau,Pv,Qv=OBJ.calc_chi_SL(Ek,phik,T0)
        
        Q=abs(Qv[0].real)
        P=Pv[0].real
        chid=chitau[0].real
        chip=chitau[1].real
        QSL_pars.append(np.array([Ui,chid,chip,P,Q]))
        print('====================================================')

filename.close()
```

```python
QSL_pars=np.array(QSL_pars)
np.save('QSL_pars.npy',QSL_pars)
```
